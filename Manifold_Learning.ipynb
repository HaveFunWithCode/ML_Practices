----------------------------------------------Concept---------------------------------------------------------------

While Linear transformers like PCA,ICA and NMF(under sklearn.decomposition module) are flexible, fast, and easily interpretable, they does not perform so well when there are nonlinear relationships within the data.

To address this deficiency, we can turn to a class of methods known as manifold learningâ€”a class of unsupervised estimators that seeks to describe datasets as low-dimensional manifolds embedded in high-dimensional spaces. these class of method(e.g Isomap, T-SNE and MDS)are available in  sklearn.manifold module.

----------------------------------------------implementation-------------------------------------------------------

<b>PCA Vs Isomap</b>

```python
%matplotlib  inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_s_curve
from mpl_toolkits.mplot3d import Axes3D

x,y = make_s_curve(n_samples=2000)
ax =plt.axes(projection='3d')
ax.scatter3D(x[:,0],x[:,1],x[:,2],c=y)
ax.view_init(20,-70)
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2).fit(x)
x_transformed=pca.transform(x)
plt.scatter(x_transformed[:,0],x_transformed[:,1],c=y)
```

In Pca some information is lost 

```python
from sklearn.manifold import Isomap

isomap=Isomap(n_neighbors=20)
isomap.fit(x)
x_transformed=isomap.transform(x)
plt.scatter(x_transformed[:,0],x_transformed[:,1],c=y)
```

<b>TSNE</b>


```python
from sklearn.manifold import TSNE
ts = TSNE()
ts.fit(x)
x_ts_transformed=ts.fit_transform(x)
```

```python
plt.scatter(x_ts_transformed[:,0],x_ts_transformed[:,1],c=y)
```

as we can see not all manifold algorithm work well on  all dataset .


<b>Test TSNE on digits dataset</b>

```python
from sklearn.datasets import load_digits
digits=load_digits(n_class=5)
x,y=digits.data,digits.target
x_ts_transformed=ts.fit_transform(x)
plt.scatter(x_ts_transformed[:,0],x_ts_transformed[:,1],c=y)
```

Tsne is good for identifying groups in the data
